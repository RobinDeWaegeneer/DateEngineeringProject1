## All flights from Brussel to Corfu, sorted by vertrekDatum.

```python
search_result = df[(df['vertrekPlaats'] == 'Brussel') & (df['aankomstPlaats'] == 'Corfu')]

sorted = search_result.sort_values(by="vertrekDatum")

print(sorted)
```
